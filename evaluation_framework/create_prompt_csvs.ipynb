```python
import os

def count_files_in_directory(directory_path):
    """Count the number of files in the given directory."""
    with os.scandir(directory_path) as dir_entries:
        return sum(1 for entry in dir_entries if entry.is_file())

directory_path = '/Users/ChaliniA/Documents/capstone-code-generation/evaluation_framework/test_10_30'
number_of_files = count_files_in_directory(directory_path)
print(f'There are {number_of_files} files in the directory {directory_path}.')
```
